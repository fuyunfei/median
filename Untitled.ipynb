```python
from scipy import misc
from scipy import ndimage
import matplotlib.pyplot as plt
f = misc.face(gray=True)
plt.imshow(f, cmap=plt.cm.gray)  

for i in range(1,10):
    m = ndimage.median_filter(f, 13)
    f=f-m


plt.imshow(f, cmap=plt.cm.gray)
plt.show()
```
